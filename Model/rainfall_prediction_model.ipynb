Data Preprocessing

df1 = df.drop('day', axis = 1)

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['winddirection'] = df1['winddirection'].fillna(df1['winddirection'].mode()[0])
df1['windspeed'] = df1['windspeed'].fillna(df1['windspeed'].median())
```

```python
df1.isnull().sum()
```

```python
df1.rainfall = df1.rainfall.map({'yes' : 1, 'no' : 0})
```
